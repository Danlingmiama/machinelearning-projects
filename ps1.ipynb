1. What is the mean and standard deviation of the numbers from -5 to 50 incremented by 3?

```python
import numpy as np
nums = np.arange(-5,51,3)
print("The numbers from -5 to 50:\n", nums,"\n")
avg_nums = np.average(nums)
avg = np.full((1, len(nums)),avg_nums)
sd_nums = (((nums - avg) ** 2).sum()*(1/len(nums)))**(1/2)
print("- The mean of the numbers from -5 to 50 incremented by 3 is", avg_nums)
print("- The standard deviation of the numbers from -5 to 50 incremented by 3 is"
      , sd_nums)
```

5. The function np.mod(a,b) in the numpy library is the modulo of a with respect to b. Ths is the remainder of a when divided by b.
    * a. Using a loop print all the numbers between 1 and 100 which are evenly divisible by 5.

```python
for num in np.arange(1,101):
    if np.mod(num, 5) == 0:
        print(num)
```

   * b. Now repeat the same procedure, but without a for loop.
   (This is a big hint. You will probably need an array of numbers between 1 and 100. 
 To do this use the command np.array(range(1,101)). A regular list will not work well. 
 You need the numpy array here.)

```python
nums = np.array(range(1,101))
modnum = nums[np.mod(nums,5) == 0]
print(modnum)
```
